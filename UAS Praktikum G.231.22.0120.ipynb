```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("stock1.csv")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 5))
plt.hist(data['Current Price'], range=(1,12))
plt.title('Distribusi data pada setiap atribut Current Price',fontsize=15, color='purple')
plt.xlabel('Current Price', fontsize = 12)
plt.ylabel('Jumlah Banyaknya', fontsize = 12)
plt.xlim(xmin=0,xmax=12)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 5))
plt.hist(data['Gain/Loss'], range=(1,12))
plt.title('Distribusi data pada setiap atribut Gain/Loss',fontsize=15, color='purple')
plt.xlabel('Gain/Loss', fontsize = 12)
plt.ylabel('Jumlah Banyaknya', fontsize = 12)
plt.xlim(xmin=0,xmax=12)
plt.show()
```

```python
import datetime
import pandas as pd
import matplotlib.pyplot as plt
data['gmv']= data['Annual Yield'] * data['Shares']
print(data.head())
```

```python
Dividend_share = (data.groupby('Stock Name')['Dividend/share']
                  .sum()
                  .reset_index()
                  .sort_values(by='Dividend/share', ascending=False)
                  .head(5))
print(Dividend_share)
```

```python
Market_Value = (data.groupby('Stock Name')['Market Value']
                .sum()
                .reset_index()
                .sort_values(by='Market Value', ascending=False)
                .head(5))
print(Market_Value)
```

```python
Annual_Yield = (data.groupby('Stock Name')['Annual Yield']
                .sum()
                .reset_index()
                .sort_values(by='Annual Yield', ascending=False)
                .head(7))
print(Annual_Yield)
```
